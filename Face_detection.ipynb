```python
# 导入必要的库
import cv2
import matplotlib.pyplot as plt
import cvlib as cv

image_file = 'couple-4445670_640.jpg'
im = cv2.imread(image_file)

plt.figure()
b,g,r=cv2.split(im)
img=cv2.merge([r,g,b])
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
faces, confidences = cv.detect_face(im)
print(faces)

# 循环遍历检测到的人脸并添加边界框
for face in faces:
    (startX,startY) = face[0],face[1]
    (endX,endY) = face[2],face[3]
    cv2.rectangle(im, (startX,startY), (endX,endY), (0,255,0), 2)  # 在人脸上绘制矩形边界框
    
# 显示结果
plt.figure()
b,g,r=cv2.split(im)
img=cv2.merge([r,g,b])
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
from scipy import misc

image_file = 'couple-4445670_640.jpg'
im0 = cv2.imread(image_file)
im1=im0

# Original faces
plt.figure()
b,g,r=cv2.split(im1)
im1_rgb=cv2.merge([r,g,b])
im1_rgb=misc.imresize(im1_rgb,2)
plt.imshow(im1_rgb)
plt.axis('off')
plt.show()

# Face detection
faces, confidences = cv.detect_face(im0)
for face in faces:
    (startX,startY,endX,endY) = face[0],face[1],face[2],face[3]
    cv2.rectangle(im0, (startX,startY), (endX,endY), (0,255,0), 2)  # 绘制矩形边界框
im2=im0

# Detected faces
plt.figure()
b,g,r=cv2.split(im2)
im2_rgb=cv2.merge([r,g,b])
im2_rgb=misc.imresize(im2_rgb,2)
plt.imshow(im2_rgb)
plt.axis('off')
plt.show()

# `imresize` is deprecated in SciPy 1.0.0, and will be removed in 1.3.0.
```
