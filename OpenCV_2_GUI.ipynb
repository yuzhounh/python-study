```python
# [OpenCV-Python] OpenCV 中的 Gui特性 部分 II - _Undo - 博客园 
# https://www.cnblogs.com/Undo-self-blog/p/8424056.html

import numpy as np
import cv2

# Load an color image in grayscale
img = cv2.imread('messi5.jpg',0)

# show image
cv2.namedWindow('image', cv2.WINDOW_NORMAL)  # 可以调整窗口大小
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# save image
cv2.imwrite('messigray.png',img)
```

```python
# 读取一幅彩色图片
# 显示为灰度图片
# 按 ESC 键退出
# 按 s 键保存后退出

img = cv2.imread('messi5.jpg',0)
cv2.imshow('image',img)
k = cv2.waitKey(0) & 0xFF
if k == 27:         # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('messigray.png',img)
    cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('messi5.jpg',0)
plt.figure()
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()

# 彩色图像使用 OpenCV 加载时是 BGR 模式。但是 Matplotib 是 RGB模式。
# 所以彩色图像如果已经被 OpenCV 读取，那它将不会被 Matplotib 正确显示。
```

```python
# 画直线

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw a diagonal blue line with thickness of 5 px
cv2.line(img,(0,0),(511,511),(255,0,0),5)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)  # 可以调整窗口大小
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 画矩形

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# draw rectangle
cv2.rectangle(img,(384,0),(510,128),(0,255,0),3)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)  # 可以调整窗口大小
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 画圆

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# draw circle
cv2.circle(img,(447,63), 63, (0,0,255), -1)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)  # 可以调整窗口大小
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 画椭圆

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# draw circle
cv2.ellipse(img,(256,256),(100,50),0,0,270,255,-1)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)  # 可以调整窗口大小
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 画多边形

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# draw polylines
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts * 6
pts = pts.reshape((-1,1,2))  # 第一个维度是由后面的两个维度计算出来的
cv2.polylines(img,[pts],True,(0,255,255))

cv2.namedWindow('image', cv2.WINDOW_NORMAL)  # 可以调整窗口大小
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 添加文字

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2,cv2.LINE_AA)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)  # 可以调整窗口大小
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
